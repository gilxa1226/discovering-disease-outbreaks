```python
import numpy
import re
import geonamescache
import pandas as pd
import unidecode

from unidecode import unidecode 

gc = geonamescache.GeonamesCache()

countries = [unidecode(name) for name in gc.get_countries_by_names().keys()]
# gc.get_countries_by_names().keys()
cities = [unidecode(value.get("name")) for key, value in gc.get_cities().items()]

def get_cities(name):
    return name in cities

def get_country(name):
    return name in countries

def create_city_name(wordlist):
    return " ".join(wordlist)

f = open("data/headlines.txt", "r", encoding="utf-8")
fileLines = f.readlines()

headline_col = []
country_col = []
city_col = []

counter = 0 

for idx in range(len(fileLines)):
    line = fileLines[idx]
    headline_col.append(line.strip())
    
    relevantWords = re.findall(r"([A-Z]\w+)\s?",line)
    partialResult = []
    city_names = []
    country_names = []
    while len(relevantWords) > 0 :
        list0 = relevantWords[:]
        
        while len(list0) > 0 :
            helperText = create_city_name(list0).strip()
            if get_cities(helperText):
                if not helperText in city_names :
                    city_names.append(helperText)
                    print(city_names)
            if(get_country(helperText)):
                country_names.append(helperText)
            list0.pop()
        relevantWords.pop(0)
    #print(city_names)
    city_names = list(filter( lambda i: len([x for x in city_names if i in x])==1 ,city_names))
    counter = counter + len(city_names) + len(country_names)
    city_col.append(", ".join(city_names) if len(city_names) > 0 else numpy.NAN  )
    country_col.append(", ".join(country_names) if len(country_names) > 0 else numpy.NAN )
    
#df = pd.DataFrame(data={'headline': headline_col, 'countries': country_col,  'cities': city_col})
#print('Cities found ', counter)
#print(df.head(50))
#df.to_csv(r'solution.csv', index=False)
```
